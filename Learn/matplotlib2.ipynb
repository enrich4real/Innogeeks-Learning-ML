```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r-+')
plt.subplot(2,2,2)
plt.plot(x,z,'g+-')
plt.subplot(2,2,3)
plt.plot(x,y,'b+-')
plt.subplot(2,2,4)
plt.plot(y,z,'m+-')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.plot(x,z,label="what")
```

```python
plt.plot(x,y)
plt.xlabel("Angles")
plt.ylabel("Sine wave")
plt.title("Sine wave")
plt.show()
```

```python
x=np.linspace(-40,40,30)
y=(x**2)
plt.plot(x,y,'r+')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph")
plt.show()
```
